# Randomly select three wt and null wells to use for IC benchmarking 

```python
import random 
random.seed(73)

# Utilized code from https://www.logilax.com/python-range-of-letters/
# Generate a list of letters in the desired range 
def range_char_list(start: str, stop: str):
    result = []
    for number in range(ord(start), ord(stop) + 1):
        result.append(chr(number))
    return result
```

Randomly select three wt wells to use 

```python
# Set range for letters 
wt_letter_range = range_char_list("b", "g")

# Get a random letter from wt_letter_range
for i in range(3):
    wt_letter = random.choice(wt_letter_range)

# make new list for results 
wt_letter_result = []

# Get a random letter from letter_range 3 x then add it to a list wt_letter_result 
for i in range(3):
    wt_letter = random.choice(wt_letter_range)
    wt_letter_result.append(wt_letter)

# make new list
wt_num_result = []

# Get random number between 2-4 and add to a list 
for i in range(3):
    wt_number = random.randint(2,4)
    wt_num_result.append(wt_number)
    
wt_well_results = []

for i in range(3):
    wt_well_results.append(wt_letter_result[i] + str(wt_num_result[i]))

print("wt wells: " + ','.join(wt_well_results))
```

Randomly select three null wells to use 

```python
# Set range for letters 
null_letter_range = range_char_list("b", "g")

# Get a random letter from null_letter_range
for i in range(3):
    null_letter = random.choice(null_letter_range)

# make new list for results 
null_letter_result = []

# Get a random letter from letter_range 3 x then add it to a list null_letter_result 
for i in range(3):
    null_letter = random.choice(null_letter_range)
    null_letter_result.append(null_letter)

# make new list
null_num_result = []

# Get random number between 9-11 and add to a list 
for i in range(3):
    null_number = random.randint(9,11)
    null_num_result.append(null_number)
    
null_well_results = []

for i in range(3):
    null_well_results.append(null_letter_result[i] + str(null_num_result[i]))

print("null wells: " + ','.join(null_well_results))
```
